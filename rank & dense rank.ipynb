```python
import pandas as pd
```

```python
emp_df = pd.DataFrame({'Name':['kevin','allwyn','dany','joe','sunil','anand'],
                  'DEPTID': [1,2,2,1,2,1],
                  'SALARY':[10000,20000,30000,60000,50000,60000]})

emp_df
```

```python
dept_df = pd.DataFrame({'DEPT_NAME':['IT','SALES'],
                       'DEPT_ID':[1,2]})
dept_df
```

```python
emp_data = pd.merge(emp_df,dept_df, how='left',\
                   left_on=['DEPTID'],right_on=['DEPT_ID'])
```

```python
emp_data
```

```python
emp_data['rank'] = emp_data.groupby(['DEPT_NAME'])['SALARY'].rank(method='dense')
```

```python
emp_data[]
```

```python
emp_data.loc[emp_data['rank'] >= 2]
```
