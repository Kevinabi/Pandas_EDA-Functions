```python
import pandas as pd
```

```python
df = pd.read_excel('EMP_SAL.xls')
df
```

```python
df.groupby('DEPT').transform('sum')
```

```python
df.groupby('DEPT').aggregate('sum')
```

```python
# Transform Groupby Object

df.groupby('DEPT').transform(lambda x: x.sum())
```

```python
# Transform Groupby Object

df.groupby('DEPT').transform('max')
```

```python
#sum of salary grouped by DEPT using Transform Func
```

```python
df.groupby(['DEPT','DESI']).transform('max')
```

```python
df.groupby('DEPT')['SALARY'].transform('sum')
```

```python
# salary on the  basis of Dept by using indexing
```

```python
df.groupby('DEPT')['SALARY'].aggregate('sum')
```

```python
# Salary on the basis of Desigation (by specified column in agg func)
```

```python
df.groupby('DESI').aggregate({'SALARY':'sum'})
```
