```python
import pandas as pd
import numpy as np
```

```python
N = 10
df = pd.DataFrame({
    'type': np.random.randint(10, size=N),
    'status': np.random.randint(10, size=N),
    'name': np.random.randint(10, size=N),
    'score': np.random.randint(10, size=N),
})
```

```python
df.sort_values(by=['type','status'])
```

```python
df.groupby(['type','status','name'],as_index=False).agg({'score': lambda x : 100 * (x.sum() * x.mean())})
```
