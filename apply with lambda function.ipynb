```python
df=pd.read_csv('mycars.csv')
df.head()

df['MSRP_adj'] = df['MSRP']
df['MSRP_adj'] = df.apply(lambda x : (x['MSRP_adj']).replace(',',''),axis=1)
df['MSRP_adj'] = df.apply(lambda x : (x['MSRP_adj']).replace('$',''),axis=1)
```

```python
df['MSRP_adj1'] = df['MSRP']

df['MSRP_adj1'] = df['MSRP_adj1'].str.replace(r',',r'',regex=True)
df['MSRP_adj1'] = df['MSRP_adj1'].str.replace(r'$',r'',regex=True)
```
